```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Load dataset
df = pd.read_csv(r"C:\Users\Administrator\Desktop\Vinuwara_IRP_A.I.D.S_2024\Code\Dataset\disdrict distribution 2020-2021.csv")

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d-%b-%Y')

# List of districts to process
districts = ['AMPARA', 'ANURADHAPURA', 'BADULLA', 'BATTICALOA', 'COLOMBO', 'GALLE', 'GAMPAHA',
             'HAMBANTOTA', 'JAFFNA', 'NUWARA ELIYA', 'POLONNARUWA', 'PUTTALAM', 'RATNAPURA',
             'TRINCOMALEE', 'VAVUNIYA']
```

```python
# SIR Model Functions
def fa(N, a, b, beta):
    return -beta * a * b

def fb(N, a, b, beta, gamma):
    return beta * a * b - gamma * b

def fc(N, b, gamma):
    return gamma * b

# SIR Model Functions
def fa(N, a, b, beta):
    return -beta * a * b

def fb(N, a, b, beta, gamma):
    return beta * a * b - gamma * b

def fc(N, b, gamma):
    return gamma * b

def rK4(N, a, b, c, fa, fb, fc, beta, gamma, hs):
    a1 = fa(N, a, b, beta) * hs
    b1 = fb(N, a, b, beta, gamma) * hs
    c1 = fc(N, b, gamma) * hs
    ak = a + a1 * 0.5
    bk = b + b1 * 0.5
    ck = c + c1 * 0.5
    a2 = fa(N, ak, bk, beta) * hs
    b2 = fb(N, ak, bk, beta, gamma) * hs
    c2 = fc(N, bk, gamma) * hs
    ak = a + a2 * 0.5
    bk = b + b2 * 0.5
    ck = c + c2 * 0.5
    a3 = fa(N, ak, bk, beta) * hs
    b3 = fb(N, ak, bk, beta, gamma) * hs
    c3 = fc(N, bk, gamma) * hs
    ak = a + a3
    bk = b + b3
    ck = c + c3
    a4 = fa(N, ak, bk, beta) * hs
    b4 = fb(N, ak, bk, beta, gamma) * hs
    c4 = fc(N, bk, gamma) * hs
    a += (a1 + 2 * (a2 + a3) + a4) / 6
    b += (b1 + 2 * (b2 + b3) + b4) / 6
    c += (c1 + 2 * (c2 + c3) + c4) / 6
    return a, b, c

def SIR(N, b0, beta, gamma, hs, ydata):
    a = float(N - 1) / N - b0
    b = float(1) / N + b0
    c = 0.

    sus, inf, rec = [], [], []
    for i in range(len(ydata)):
        sus.append(a)
        inf.append(b)
        rec.append(c)
        a, b, c = rK4(N, a, b, c, fa, fb, fc, beta, gamma, hs)

    return sus, inf, rec
def SIR(N, b0, beta, gamma, hs, ydata):
    a = float(N - 1) / N - b0
    b = float(1) / N + b0
    c = 0.
    sus, inf, rec = [], [], []
    for i in range(len(ydata)):
        sus.append(a)
        inf.append(b)
        rec.append(c)
        a, b, c = rK4(N, a, b, c, fa, fb, fc, beta, gamma, hs)
    return sus, inf, rec

# Model Parameters
N = 23229470  # SL population
b0 = 0
beta = 0.3
gamma = 0.1
hs = 0.1
```

```python
import os

# Create a directory to save the plots if it doesn't exist
save_dir = "regression_plots_reigons"
os.makedirs(save_dir, exist_ok=True)

for district in districts:
    if district in df.columns:
        print(f"Processing {district}...")
        
        # Extract data for the district
        district_df = df[['Date', district]].dropna()
        district_df = district_df[district_df[district] > 0]
        district_df = district_df.sort_values(by='Date')
        
        # Convert dates to numeric for regression
        district_df['Days_Since_Start'] = (district_df['Date'] - district_df['Date'].min()).dt.days
        X = district_df[['Days_Since_Start']].values
        Y = district_df[district].values
        
        # Linear Regression Model
        linear_model = LinearRegression()
        linear_model.fit(X, Y)
        linear_pred = linear_model.predict(X)
        
        # Polynomial Regression Model
        poly = PolynomialFeatures(degree=5)
        X_poly = poly.fit_transform(X)
        poly_model = LinearRegression()
        poly_model.fit(X_poly, Y)
        poly_pred = poly_model.predict(X_poly)
        
        # Logarithmic Regression Model
        log_Y = np.log(Y + 1)
        log_model = LinearRegression()
        log_model.fit(X, log_Y)
        log_pred = np.exp(log_model.predict(X)) - 1
        
        # Apply SIR Model
        _, inf, _ = SIR(N, b0, beta, gamma, hs, Y)
        
        # Plot results
        plt.figure(figsize=(18, 6))
        
        # Linear Regression Plot
        plt.subplot(1, 3, 1)
        plt.scatter(X, Y, color='blue', label='Actual Data')
        plt.plot(X, linear_pred, color='red', label='Linear Regression')
        plt.xlabel("Days Since Start")
        plt.ylabel("Total Infected Cases")
        plt.title(f"Linear Regression - {district}")
        plt.legend()
        
        # Polynomial Regression Plot
        plt.subplot(1, 3, 2)
        plt.scatter(X, Y, color='blue', label='Actual Data')
        plt.plot(X, poly_pred, color='purple', label='Polynomial Regression')
        plt.xlabel("Days Since Start")
        plt.ylabel("Total Infected Cases")
        plt.title(f"Polynomial Regression - {district}")
        plt.legend()
        
        # Logarithmic Regression Plot
        plt.subplot(1, 3, 3)
        plt.scatter(X, Y, color='blue', label='Actual Data')
        plt.plot(X, log_pred, color='green', label='Log Regression')
        plt.xlabel("Days Since Start")
        plt.ylabel("Total Infected Cases")
        plt.title(f"Logarithmic Regression - {district}")
        plt.legend()
        
        plt.tight_layout()
        plt.show()
        
        # Save the figure
        save_path = os.path.join(save_dir, f"{district}_regression.png")
        plt.savefig(save_path, dpi=300)
        plt.close()  # Close the figure to free up memory
        
        print(f"Graphs saved for {district} at {save_path}")
        
        # Print Model Coefficients
        print(f"\nLinear Regression Coefficients for {district}:")
        print("Slope:", linear_model.coef_[0], "Intercept:", linear_model.intercept_)
        print("\nPolynomial Regression Coefficients:")
        print("Coefficients:", poly_model.coef_, "Intercept:", poly_model.intercept_)
        print("\nLogarithmic Regression Coefficients:")
        print("Slope:", log_model.coef_[0], "Intercept:", log_model.intercept_)
        
        print("--------------------------------------------------")

        
```

```python
# Evaluation function for model performance
def evaluate_model(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    return mae, mse, rmse, r2

for district in districts:
    if district in df.columns:
        print(f"Processing {district}...")
        
        # Extract data for the district
        district_df = df[['Date', district]].dropna()
        district_df = district_df[district_df[district] > 0]
        district_df = district_df.sort_values(by='Date')
        
        # Convert dates to numeric for regression
        district_df['Days_Since_Start'] = (district_df['Date'] - district_df['Date'].min()).dt.days
        X = district_df[['Days_Since_Start']].values
        Y = district_df[district].values
        
        # Linear Regression Model
        linear_model = LinearRegression()
        linear_model.fit(X, Y)
        linear_pred = linear_model.predict(X)
        
        # Polynomial Regression Model
        poly = PolynomialFeatures(degree=5)
        X_poly = poly.fit_transform(X)
        poly_model = LinearRegression()
        poly_model.fit(X_poly, Y)
        poly_pred = poly_model.predict(X_poly)
        
        # Logarithmic Regression Model
        log_Y = np.log(Y + 1)
        log_model = LinearRegression()
        log_model.fit(X, log_Y)
        log_pred = np.exp(log_model.predict(X)) - 1
        
        # Evaluate the models
        print(f"\nEvaluating models for {district}:")
        
        # Linear Regression Evaluation
        linear_mae, linear_mse, linear_rmse, linear_r2 = evaluate_model(Y, linear_pred)
        print(f"Linear Regression Metrics:")
        print(f"MAE: {linear_mae}, MSE: {linear_mse}, RMSE: {linear_rmse}, R: {linear_r2}")
        
        # Polynomial Regression Evaluation
        poly_mae, poly_mse, poly_rmse, poly_r2 = evaluate_model(Y, poly_pred)
        print(f"Polynomial Regression Metrics:")
        print(f"MAE: {poly_mae}, MSE: {poly_mse}, RMSE: {poly_rmse}, R: {poly_r2}")
        
        # Logarithmic Regression Evaluation
        log_mae, log_mse, log_rmse, log_r2 = evaluate_model(Y, log_pred)
        print(f"Logarithmic Regression Metrics:")
        print(f"MAE: {log_mae}, MSE: {log_mse}, RMSE: {log_rmse}, R: {log_r2}")
        
        print("--------------------------------------------------")
```
